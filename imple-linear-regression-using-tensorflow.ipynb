```python
# Import standard libraries
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

- **def generate_data:** **"def"** sirve para la definicion de funciones donde en este caso es **"generate_data"** con tres parametros los cuales son los siguientes:
  
    - **true_coeffs=np.array([10, 1]):** Se iguala el valor de un arreglo (1 renglon * 2 columnas) con los datos especificados dentro del mismo a **"true_coeffs"**.
    - **noise_std=2:** Desviación estándar.
    - **num_samples=100:** Numero de muestras a obtener.
      
- **np.random.seed (4500):** np.random implementa numeros aleatorios con la capacidad de extraer muestras de una variedad de distribuciones de probabilidad. La **"seed"** se refiere a números enteros positivos grandes donde se recomienda utilizar numeros únicos y grandes para garantizar que la semilla sea diferente a la de cualquier otra persona. "This is good practice to ensure that your results are statistically independent from theirs unless you are intentionally trying to reproduce their result. A convenient way to get such a seed number is to use secrets.randbits to get an arbitrary 128-bit integer"(Random Sampling (numpy.random) — NumPy v2.1 Manual, n.d.).
  
- **x=np.arange(0,100,1):** Crea una matriz en el rango definido de principio (0) a fin (100) y en pasos intermedios/aumento de (1).
- **y:** Función conformada por los parametros siguientes:
  
    - **true_coeffs[0]:** Parametro de la ecuación lineal **"y"**. Este coeficiente fue definido en la linea **true_coeffs=np.array([10, 1]):** donde la posicion/columna **"0"** es **"10"**.
    - **true_coeffs[1]:** Parametro de la ecuación lineal **"y"**. Este coeficiente fue definido en la linea **true_coeffs=np.array([10, 1]):** donde la posicion/columna **"1"** es **"1"**.
    - **np.random.normal(loc=0.0, scale=noise_std, size=num_samples):** Se usa **"normal"** porque se crea una matriz de números aleatorios siguiendo esta distribución (Gaussiana) donde:
      
    - **loc=0.0:** Media de la distribución.
    - **scale=noise_std:** La desviación estándar (extensión o ancho) de la distribución.
    - **size=num_samples:** La forma de la matriz de salida. Si no se proporciona, se devuelve un valor único.
      
-  **return (x, y):** Imprime el resultado de la función.

```python
def generate_data(true_coeffs=np.array([10, 1]), noise_std=2, num_samples=100):

    # Seed for reproducibility
    np.random.seed (4500)

    # Generate synthetic data
    x = np.arange(0,100, 1)
    y = true_coeffs [0] + true_coeffs[1]*x + np.random.normal(loc=0.0, scale=noise_std, size=num_samples)

    # Return features and response
    return (x, y)
```

- **true_coeffs = np.array([10, 1]):** Se iguala el valor de un arreglo (1 renglon * 2 columnas) con los datos especificados dentro del mismo a **"true_coeffs"**.

- **noise_std = 5:** Desviación estándar.

- **x, y = generate_data(true_coeffs=true_coeffs, noise_std=noise_std, num_samples=100):** Uso de función de **"generate_data"** para la regresión lineal.


```python
# Generate a synthetic dataset for simple linear regression
true_coeffs = np.array([10, 1])
noise_std = 5
x, y = generate_data(true_coeffs=true_coeffs, noise_std=noise_std, num_samples=100)
```

- **model = tf.keras.Sequential:** Creación de un modelo secuencial de Keras.

- **tf.keras.layers.Input(shape=(1,)):** Añade una capa de entrada al modelo, el modelo espera una entrada.

- **tf.keras.layers.Dense(1):** Añade una capa densa con una neurona al modelo. Como salida, la predicción de la regresión.

- **model.compile(optimizer='adam', loss='mse'):** Compila el modelo antes de entrenarlo.  
  - **optimizer='adam':** Método de descenso de gradiente estocástico que se basa en la estimación adaptativa de momentos de primer y segundo orden.  
  - **loss='mse':** Función de pérdida como el error cuadrático medio.

- **model.summary():** Muestra un resumen del modelo.

- **model.fit(x_train, y_train, epochs=500, batch_size=10, validation_split=0.10, verbose=1):** Entrena el modelo con `x_train` y `y_train`.  
  - **epochs=500:** Número de iteraciones que el modelo realiza para su entrenamiento.  
  - **batch_size=10:** Tamaño de los lotes en los que se dividen los datos.  
  - **validation_split=0.10:** Separa el 10% de los datos de entrenamiento para usarlos como validación.  
  - **verbose=1:** Muestra cada iteración y su avance.

- **loss = model.evaluate(x_test, y_test):** Evaluar el modelo para calcular la pérdida..

- **print():** Imprime la pérdida obtenida.

```python
# Build the linear regression model using a multiple-input single neuron
model = tf. keras. Sequential ( [
tf.keras. layers. Input(shape=(1,)),
tf.keras. layers. Dense(1)
])
# Compile the model
model. compile(optimizer='adam', loss='mse')

model.summary()

# Train the model
model. fit(x_train, y_train, epochs=500, batch_size=10, validation_split=0.10, verbose=0) ##verbose= 1

#Le puse verbose=0 porque así no se hace tan largo el PDF

# Evaluate the model
loss = model.evaluate(x_test, y_test)
print(f'\nTest Loss: {loss} ')
```
